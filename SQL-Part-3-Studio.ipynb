### **Event 1:** Women's History Month

Goal: Highlight Popular Womans Authors based on ratings from the Good Reads DB.

### **Write a query that returns tag\_id, the number of times each tag\_id is used and the tag\_name.**  

### Use the GROUP BY and HAVING clause to narrow your focus.

Try multiple keywords, ie woman vs female (female ended up giving better results)

```sql
-- Solution
SELECT bt.tag_id, COUNT(bt.tag_id) as tag_count, t.tag_name
FROM BooksDB.dbo.book_tags AS bt
JOIN BooksDB.dbo.tags AS t
ON bt.tag_id = t.tag_id
WHERE t.tag_name LIKE '%female%' OR t.tag_name LIKE '%woman%'
GROUP BY bt.tag_id, t.tag_name
HAVING COUNT(bt.tag_id) > 100
ORDER BY COUNT(bt.tag_id) DESC;
```

### **Write a query to return authors, titles, ratings and** **selected tag(s)\_id.**

```sql
-- Solution
SELECT b.authors, COUNT(DISTINCT b.best_book_id) as num_books, AVG(b.average_rating) as avg_rating, SUM(b.work_ratings_count) as total_reviews
FROM BooksDB.dbo.books AS b
JOIN BooksDB.dbo.book_tags AS bt
ON b.best_book_id = bt.goodreads_book_id
WHERE bt.tag_id IN (
    SELECT t.tag_id
    FROM BooksDB.dbo.tags AS t
    WHERE t.tag_name LIKE '%female%' OR t.tag_name LIKE '%woman%'
)
GROUP BY b.authors
HAVING COUNT(DISTINCT b.best_book_id) > 3 AND AVG(b.average_rating) > 4
ORDER BY total_reviews DESC;
```

```sql
-- Event 1 Query Unlock Reality with Nonfiction November!
SELECT b.title, b.average_rating, COUNT(bt.tag_id) as tag_count, t.tag_name
FROM BooksDB.dbo.books AS b
JOIN BooksDB.dbo.book_tags AS bt ON b.best_book_id = bt.goodreads_book_id
JOIN BooksDB.dbo.tags AS t ON bt.tag_id = t.tag_id
WHERE t.tag_name LIKE '%nonfiction%'
GROUP BY b.title, t.tag_name, b.average_rating
ORDER BY b.average_rating DESC, tag_count DESC;
```
